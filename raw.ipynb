```python
import json
import pandas as pd
import numpy as np
```

```python
# reading the JSON data using json.load()
file = 'sample_data/sample.json'
with open(file) as f:
    shopee_json = json.load(f)

# covert into string
# data = json.dumps(shopee_json,indent=2)
# print(data)
# print('items\titemid\twelcome_package_info\tliked\trecommendation_info')
itemid_arr = []
name_arr = []
raw_discount_arr = []
liked_count_arr = []
recommendation_info_arr = []
price_arr = []
shopid_arr = []
historical_sold_arr = []

for item in shopee_json['items']:
  itemid = item['itemid']
  name = item['name']
  raw_discount = item['raw_discount']
  liked_count = item['liked_count']
  recommendation_info = item['recommendation_info']
  price = item['price']
  shopid = item['shopid']
  historical_sold = item['historical_sold']
  # print(itemid,welcome_package_info,liked,recommendation_info,price_max_before_discount)
  itemid_arr.append(itemid)
  name_arr.append(name)
  raw_discount_arr.append(raw_discount)
  liked_count_arr.append(liked_count)
  recommendation_info_arr.append(recommendation_info)
  price_arr.append(price)
  shopid_arr.append(shopid)
  historical_sold_arr.append(historical_sold)


df = pd.DataFrame()
df['itemid']=itemid_arr
df['name']=name_arr
df['raw_discount'] = raw_discount_arr
df['liked_count'] = liked_count_arr
# df['recommendation_info'] = recommendation_info_arr
df['price'] = price_arr
df['shopid'] = shopid_arr
df['historical_sold'] = historical_sold_arr
print((df['price'].dtype))
df
```
